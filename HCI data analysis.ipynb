```python
import matplotlib.pyplot as plt
import numpy as np
import json

#load data into json 
f = open('mydata.json')
data = json.load(f)

experiment_number = 1
experiment1 = data["1"]
experiment2 = data["2"]
experiments = [experiment1,experiment2]
difficulty_levels = sorted(set(int(key.split('-')[0])for key in experiment1))
distance_combinations = sorted(set(int(key.split('-')[1])for key in experiment1))

for experiment in experiments:
    plt.figure(figsize = (10,6))
    idtimes = []
    for difficulty_level in difficulty_levels:
        times = []
        for distance_combination in distance_combinations:
            key = f"{difficulty_level}-{distance_combination}"
            times.append(np.mean(experiment[key]))
        idtimes.append(np.mean(times))
    plt.plot(difficulty_levels,idtimes,marker="o", label=f"Experiment 1")
    plt.xlabel('Difficulty Index')
    plt.ylabel('Average time (ms)')
    plt.title(f'Fitts Law Experiment {experiment_number} Results')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'fittslaw{experiment_number}.png')
    plt.show()
    
    experiment_number = experiment_number+1
    

                        
```
